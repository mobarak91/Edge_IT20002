Plotting Hyperbola

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the hyperbola
a = 2
b = 1

# Generate values for x
x = np.linspace(-5, 5, 400)

# Calculate corresponding y values for the upper and lower branches of the hyperbola
y_upper = b * np.sqrt((x**2 / a**2) - 1)
y_lower = -b * np.sqrt((x**2 / a**2) - 1)

# Plot the hyperbola branches
plt.plot(x, y_upper, 'r', label=r'$\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$')
plt.plot(x, y_lower, 'r')

# Add asymptotes for reference
asymptote_y1 = (b / a) * x
asymptote_y2 = -(b / a) * x
plt.plot(x, asymptote_y1, 'g--', label='Asymptotes')
plt.plot(x, asymptote_y2, 'g--')

# Setting the plot limits
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title("Plot of the Hyperbola")
plt.legend()
plt.grid(True)
plt.show()
```
