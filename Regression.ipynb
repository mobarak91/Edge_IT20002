Practicing Scipy, Matplotlib, Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Generate synthetic data
np.random.seed(0)  # for reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = 3 * x + 7 + np.random.normal(0, 3, size=x.size)  # y = 3*x + 7 + noise

# Step 2: Perform linear regression using SciPy
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print regression results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")

# Step 3: Plot data and regression line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color="blue", label="Data points")
plt.plot(x, slope * x + intercept, color="red", label=f"Regression line: y={slope:.2f}x + {intercept:.2f}")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression with SciPy and Matplotlib")
plt.legend()
plt.show()
```
